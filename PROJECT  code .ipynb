**#### Random Forest ####**

```python
# ==================================================
# Step 1: Upload Dataset
# ==================================================

from google.colab import files
uploaded = files.upload()

# ==================================================
# Step 2: Import Libraries
# ==================================================

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

# Fix graph spacing globally
plt.rcParams["figure.autolayout"] = True


# ==================================================
# Step 3: Load Dataset
# ==================================================

df = pd.read_csv("Crop_recommendation.csv")

print("Dataset Loaded Successfully\n")

print(df.head())


# ==================================================
# Step 4: Bar Chart (Crop Distribution)
# ==================================================

plt.figure(figsize=(10,6))

df["label"].value_counts().plot(kind="bar")

plt.title("Crop Distribution (Bar Chart)")

plt.xlabel("Crop Type")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


# ==================================================
# Step 5: Pie Chart (Crop Distribution)
# ==================================================

plt.figure(figsize=(8,8))

df["label"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Crop Distribution (Pie Chart)")

plt.ylabel("")

plt.tight_layout()

plt.show()


# ==================================================
# Step 6: Data Preprocessing
# ==================================================

le = LabelEncoder()

df["label"] = le.fit_transform(df["label"])


# ==================================================
# Step 7: Split Data
# ==================================================

X = df.drop("label", axis=1)

y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)


# ==================================================
# Step 8: Train Random Forest Model
# ==================================================

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)


# ==================================================
# Step 9: Prediction
# ==================================================

y_pred = model.predict(X_test)


# ==================================================
# Step 10: Evaluation Metrics
# ==================================================

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(
    y_test,
    y_pred,
    average="weighted"
)

f1 = f1_score(
    y_test,
    y_pred,
    average="weighted"
)

print("\n Random Forest Results")

print("Accuracy :", round(accuracy * 100, 2), "%")

print("Precision :", round(precision * 100, 2), "%")

print("F1 Score :", round(f1 * 100, 2), "%")


# ==================================================
# Step 11: Metrics Bar Chart
# ==================================================

plt.figure(figsize=(6,4))

metrics = ["Accuracy", "Precision", "F1 Score"]

values = [accuracy, precision, f1]

plt.bar(metrics, values)

plt.title("Random Forest Performance Metrics")

plt.ylabel("Score")

plt.tight_layout()

plt.show()


# ==================================================
# Step 12: Show Input Ranges
# ==================================================

print("\n========== INPUT RANGES ==========\n")

for column in X.columns:

    print(
        f"{column} : "
        f"{df[column].min()} to {df[column].max()}"
    )


# ==================================================
# Step 13: User Input
# ==================================================

print("\nEnter Soil Details (within range)\n")

N = float(input("Enter Nitrogen (N): "))

P = float(input("Enter Phosphorus (P): "))

K = float(input("Enter Potassium (K): "))

temperature = float(
    input("Enter Temperature (°C): ")
)

humidity = float(
    input("Enter Humidity (%): ")
)

ph = float(
    input("Enter pH Value: ")
)

rainfall = float(
    input("Enter Rainfall (mm): ")
)


# ==================================================
# Step 14: Prediction from User Input
# ==================================================

user_data = pd.DataFrame([{

    "N": N,
    "P": P,
    "K": K,
    "temperature": temperature,
    "humidity": humidity,
    "ph": ph,
    "rainfall": rainfall

}])

prediction = model.predict(user_data)

crop_name = le.inverse_transform(prediction)

print("\n Recommended Crop:")

print(crop_name[0])
```

```python
# ==================================================
# Step 1: Upload Dataset
# ==================================================

from google.colab import files
uploaded = files.upload()

# ==================================================
# Step 2: Import Libraries
# ==================================================

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

# Fix graph spacing globally
plt.rcParams["figure.autolayout"] = True


# ==================================================
# Step 3: Load Dataset
# ==================================================

df = pd.read_csv("Crop_recommendation.csv")

print("Dataset Loaded Successfully\n")

print(df.head())


# ==================================================
# Step 4: Bar Chart (Crop Distribution)
# ==================================================

plt.figure(figsize=(10,6))

df["label"].value_counts().plot(kind="bar")

plt.title("Crop Distribution (Bar Chart)")

plt.xlabel("Crop Type")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


# ==================================================
# Step 5: Pie Chart (Crop Distribution)
# ==================================================

plt.figure(figsize=(8,8))

df["label"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Crop Distribution (Pie Chart)")

plt.ylabel("")

plt.tight_layout()

plt.show()


# ==================================================
# Step 6: Data Preprocessing
# ==================================================

le = LabelEncoder()

df["label"] = le.fit_transform(df["label"])


# ==================================================
# Step 7: Split Data
# ==================================================

X = df.drop("label", axis=1)

y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)


# ==================================================
# Step 8: Train Decision Tree Model
# ==================================================

model = DecisionTreeClassifier(
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)


# ==================================================
# Step 9: Prediction
# ==================================================

y_pred = model.predict(X_test)


# ==================================================
# Step 10: Evaluation Metrics
# ==================================================

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(
    y_test,
    y_pred,
    average="weighted"
)

f1 = f1_score(
    y_test,
    y_pred,
    average="weighted"
)

print("\n Decision Tree Results")

print("Accuracy :", round(accuracy * 100, 2), "%")

print("Precision :", round(precision * 100, 2), "%")

print("F1 Score :", round(f1 * 100, 2), "%")


# ==================================================
# Step 11: Metrics Bar Chart
# ==================================================

plt.figure(figsize=(6,4))

metrics = ["Accuracy", "Precision", "F1 Score"]

values = [accuracy, precision, f1]

plt.bar(metrics, values)

plt.title("Decision Tree Performance Metrics")

plt.ylabel("Score")

plt.tight_layout()

plt.show()


# ==================================================
# Step 12: Show Input Ranges
# ==================================================

print("\n========== INPUT RANGES ==========\n")

for column in X.columns:

    print(
        f"{column} : "
        f"{df[column].min()} to {df[column].max()}"
    )


# ==================================================
# Step 13: User Input
# ==================================================

print("\nEnter Soil Details (within range)\n")

N = float(input("Enter Nitrogen (N): "))

P = float(input("Enter Phosphorus (P): "))

K = float(input("Enter Potassium (K): "))

temperature = float(
    input("Enter Temperature (°C): ")
)

humidity = float(
    input("Enter Humidity (%): ")
)

ph = float(
    input("Enter pH Value: ")
)

rainfall = float(
    input("Enter Rainfall (mm): ")
)


# ==================================================
# Step 14: Prediction from User Input
# ==================================================

user_data = pd.DataFrame([{

    "N": N,
    "P": P,
    "K": K,
    "temperature": temperature,
    "humidity": humidity,
    "ph": ph,
    "rainfall": rainfall

}])

prediction = model.predict(user_data)

crop_name = le.inverse_transform(prediction)

print("\n Recommended Crop:")

print(crop_name[0])
```
